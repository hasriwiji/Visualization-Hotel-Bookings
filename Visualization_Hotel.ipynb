```python
X = data3.iloc[:, [11,14]].values

from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
    kmeans = KMeans (n_clusters = i, init = 'k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.ylabel('wcss')
plt.xlabel('no of clusters')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3).fit(data3)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(data3['is_repeated_guest'],data3['booking_changes'], c=kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:,0],centroids[:,1], c='red',s=100, label = 'Cluster 1')
plt.scatter(centroids[:,0],centroids[:,1], c='pink',s=50, label = 'Cluster 2')
plt.scatter(centroids[:,0],centroids[:,1], c='yellow',s=40, label = 'Cluster 3')
plt.title('Clusters of Houses to Rent')
plt.xlabel('is_repeated_guest	')
plt.ylabel('booking_changes')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

standard_scaler = StandardScaler()
new_scaled=standard_scaler.fit_transform(data3)
```

```python
pca=PCA(svd_solver='randomized',random_state=42)
pca.fit(new_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
fig=plt.figure(figsize=(10,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
colnames = list (data3.columns)
pcs_data3 = pd.DataFrame({'Feature':colnames,'PC1':pca.components_[0],'PC2':pca.components_[1],'PC3':pca.components_[2],'PC4':pca.components_[3]
                         ,'PC5':pca.components_[4],'PC6':pca.components_[5],'PC7':pca.components_[6],'PC8':pca.components_[7],
                         'PC9':pca.components_[8],'PC10':pca.components_[9]})
pcs_data3
```
